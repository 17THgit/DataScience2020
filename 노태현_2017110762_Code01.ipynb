```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/birds.csv') #구글 드라이브에서 csv 파일 읽어옴
print(df) #csv 파일 데이터 출력
pd.value_counts(df['type']).plot.pie() #type에 따라 비율을 알기 쉽게 하기 위해 데이터를 Pie plot으로
plt.show() #출력
```

```
df.plot.hist(subplots=True, layout=(2, 5), alpha=1.0, bins=100, figsize=(20, 10)) #type에 상관 없이 데이터에서의 빈도수 및 이상치 관찰을 위해 histogram
plt.show() #출력
```

```
m=df.mean() #type에 상관없이 전체 데이터들에 대한 평균
print(m) #출력
```

```
m_p=df[df['type']=='P'].mean() #type으로 P를 가지는 데이터들에 대한 평균
print(m_p) #출력
c_p = df[df['type']=='P'].cov() #type으로 P를 가지는 데이터들에 대한 공분산
print(c_p) #출력
```
