```
from google.colab import drive
drive.mount('/content/gdrive')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/birds.csv') #구글 드라이브에서 csv 파일 읽어옴
print(df) #csv 파일 데이터 출력
```

```
# linear regression - humerus length, humerus width 비교

from sklearn.linear_model import LinearRegression

x = df["huml"]
y = df["humw"]
N = x.shape[0]

plt.figure(figsize=(15, 10))
plt.title("humerus length / humerus width")
plt.xlabel("humerus length (mm)")
plt.ylabel("humerus width (mm)")

plt.scatter(x,y)  # x, y 데이터들의 산점도 출력

mean_x = np.mean(x) # x의 평균
mean_y = np.mean(y) # y의 평균

w1 = (N*np.sum(x*y) - np.sum(x)*np.sum(y))/(N*np.sum(x*x)-np.sum(x)**2) # 선형회귀방정식의 기울기를 구하는 수식
w0 = mean_y - w1*mean_x # 선형회귀방정식의 y절편을 구하는 수식
print('w1:', w1,', w2:', w0,'이므로')
print('y =', w1,'x +',w0) # 선형회귀방정식 출력
# Calculate predicted y values using fitted line
predict_y = w1*x +w0  # x를 대입했을 때 나오는 예측치 predict_y
plt.scatter(x,predict_y,c='r')  # 대입치, 예측치의 산점도를 통해 얻는 그래프 출력
# Calculate RMS error
RMSE = np.sqrt(np.mean((predict_y - y)**2)) # RMS 구하기
print('mean:', mean_y,'RMSE:',RMSE)
print('변동성:', (RMSE/mean_y)*100,'%') # 예측치와 실제 값이 퍼져있는 정도인 변동성 구하기
#predict
```

```
# std length dendrogram

from scipy.cluster.hierarchy import linkage, dendrogram

huml = df["huml"]
ulnal = df["ulnal"]
feml = df["feml"]
tibl = df["tibl"]
tarl = df["tarl"]

std_huml = np.std(huml) # humerus length 데이터들의 표준편차 구하기
std_ulnal = np.std(ulnal) # ulna length 데이터들의 표준편차 구하기
std_feml = np.std(feml) # femur length 데이터들의 표준편차 구하기
std_tibl = np.std(tibl) # tibiotarsus length 데이터들의 표준편차 구하기
std_tarl = np.std(tarl) # tarsometatarsus length 데이터들의 표준편차 구하기

print('std_huml:', std_huml,'std_ulnal:', std_ulnal,'std_feml:', std_feml,'std_tibl:', std_tibl,'std_tarl:', std_tarl)

Xs = [[std_huml], [std_ulnal], [std_feml], [std_tibl], [std_tarl]]  # 각 뼈의 길이에 대한 표준편차들을 데이터에 넣기

Zs = linkage(Xs, 'single')  # 단일연결법을 이용한 덴드로그램 그리기
plt.figure(figsize=(15, 10))
dendrogram(Zs)
plt.title("std's single dendrogram by length of bones")
plt.show()
print('        0 = std_huml, 1 = std_ulnal, 2 = std_feml, 3 = std_tibl, 4 = std_tarl')
print()
print(Zs)

Zs = linkage(Xs, 'average') # 평균연결법을 이용한 덴드로그램 그리기
plt.figure(figsize=(15, 10))
dendrogram(Zs)
plt.title("std's average dendrogram by length of bones")
plt.show()
print('        0 = std_huml, 1 = std_ulnal, 2 = std_feml, 3 = std_tibl, 4 = std_tarl')
print()
print(Zs) 

Zs = linkage(Xs, 'complete')  # 최장연결법을 이용한 덴드로그램 그리기
plt.figure(figsize=(15, 10))
dendrogram(Zs)
plt.title("std's complete dendrogram by length of bones")
plt.show()
print('        0 = std_huml, 1 = std_ulnal, 2 = std_feml, 3 = std_tibl, 4 = std_tarl')
print()
print(Zs) 
```
